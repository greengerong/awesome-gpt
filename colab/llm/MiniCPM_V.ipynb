# 推理

````python
import gc
gc.collect()
torch.cuda.empty_cache()
````

````python
%cd /content/MiniCPM-V
!wget https://huggingface.co/openbmb/MiniCPM-V-2.0/resolve/main/assets/minicpmv-2-benchmark.png -O minicpmv-2-benchmark.png

import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

model = AutoModel.from_pretrained('openbmb/MiniCPM-V-2', trust_remote_code=True, torch_dtype=torch.float16)
model = model.to(device='cuda', dtype=torch.float16)

tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-V-2', trust_remote_code=True)
model.eval()
````

````python
%cd /content/MiniCPM-V

image = Image.open('./minicpmv-2-benchmark.png').convert('RGB')
# question = 'Please list all the performance score with table. Forst column is model name, and all of the measure score.'
question = '请问图片中内容是什么？'
msgs = [{'role': 'user', 'content': question}]

res, context, _ = model.chat(
    image=image,
    msgs=msgs,
    context=None,
    tokenizer=tokenizer,
    sampling=True,
    temperature=0.7
)
print(res)
````
